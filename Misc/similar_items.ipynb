```python
# Functions to get_title_from_index and get_index_from_title
def get_title_from_index(index):
    return df1.iloc[index]["ITEM_NAME"]

def get_index_from_title(title):
    return pd.DataFrame(df1.loc[df1['ITEM_NAME'] == title])["index"].values[0]


def get_index_from_id(item_id):
    return pd.DataFrame(df1.loc[df1['ITEM_ID'] == item_id])["index"].values[0]

def get_id_from_index(index):
    return df1.iloc[index]["ITEM_ID"]
```
